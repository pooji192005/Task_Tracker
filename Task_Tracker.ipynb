```python
tasks = []
while True:
    print("\n1.Add Task")
    print("2.View Task")
    print("3.Exit")
    choice = input("Choose option:")
    if choice == "1":
        task = input("Enter task:")
        tasks.append(task)
        print("Task Added!")
    elif choice == "2":
        for i,task in enumerate(tasks):
            print(i+1,task)
    elif choice == "3":
        break
    else :
        print("Invalid Choice")
```
